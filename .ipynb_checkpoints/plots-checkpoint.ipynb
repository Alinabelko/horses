```python
#!/usr/bin/env python3
import ptan
import pathlib
import argparse
import gym.wrappers
import numpy as np

import torch
import torch.optim as optim

from ignite.engine import Engine
from ignite.contrib.handlers import tensorboard_logger as tb_logger

from lib import environ, data, models, common, validation
import matplotlib.pyplot as plt
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"


SAVES_DIR = pathlib.Path("/saves")
STOCKS = "data/YNDX_160101_161231.csv"
VAL_STOCKS = "data/YNDX_150101_151231.csv"

FOR_DATA = VAL_STOCKS
FOR_MODELS = "saves/test/mean_val-0.462.data"
BATCH_SIZE = 32
BARS_COUNT = 10

EPS_START = 1.0
EPS_FINAL = 0.1
EPS_STEPS = 1000000

GAMMA = 0.99

REPLAY_SIZE = 100000
REPLAY_INITIAL = 10000
REWARD_STEPS = 2
LEARNING_RATE = 0.0001
STATES_TO_EVALUATE = 1000

EPSILON = 0.02

if __name__ == "__main__":
    o_in = {'data' :FOR_DATA ,'model' : FOR_MODELS,'bars' : 10,'name' :'test', 'commission':0.0,'conv':False}

    prices = data.load_relative(o_in['data'])
    env = environ.StocksEnv({"TEST": prices}, bars_count=o_in['bars'], reset_on_close=False, commission=o_in['commission'],
                            state_1d=o_in['conv'], random_ofs_on_reset=False, reward_on_close=False, volumes=False)
    if o_in['conv']:
        net = models.DQNConv1D(env.observation_space.shape, env.action_space.n)
    else:
        net = models.SimpleFFDQN(env.observation_space.shape[0], env.action_space.n)

    net.load_state_dict(torch.load(o_in['model'], map_location=lambda storage, loc: storage))

    obs = env.reset()
    start_price = env._state._cur_close()

    total_reward = 0.0
    step_idx = 0
    rewards = []

    while True and step_idx < 20000:
        step_idx += 1
        obs_v = torch.tensor([obs])
        out_v = net(obs_v)
        action_idx = out_v.max(dim=1)[1].item()
        if np.random.random() < EPSILON:
            action_idx = env.action_space.sample()
        action = environ.Actions(action_idx)

        obs, reward, done, _ = env.step(action_idx)
        total_reward += reward
        rewards.append(total_reward)
        if step_idx % 100 == 0:
            print("%d: reward=%.3f" % (step_idx, total_reward))
        if done:
            break

    plt.clf()
    plt.plot(rewards)
    plt.title("Total reward, data=%s" % o_in['name'])
    plt.ylabel("Reward, %")
    plt.savefig("rewards-%s.png" % o_in['name'])
```
